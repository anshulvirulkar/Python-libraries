```python
# Q.4. Does price varies with airlines ?

data.head()
```

```python
# Grouping the airlines and checking their mean price

data.groupby('airline')['price'].mean()
```

```python
# Drawing a Categorical Plot showing the Mean Ticket Price for each Airline

# catplot() function to create a categorical bar plot comparing flight prices across different airlines, 
# segmented by class (Economy/Business)

sns.catplot(
    x='airline',        # Categorical variable on X-axis (e.g., airline names)
    y='price',          # Numerical variable on Y-axis (flight prices)
    kind='bar',         # Type of plot: bar chart
    palette='dark',    # Color scheme (other default palette- dark, bright, deep, pastel, colorblind(seaborn default)
    data=data,          # DataFrame containing the data
    hue='class'         # Sub-grouping by 'class' (e.g., Economy/Business)
)
plt.show()
```

```python
# Q.5. Does ticket price change based on the departure time and arrival time?

data.head()
```

```python
data.groupby('departure_time')['price'].mean()
```

```python
# Checking the Mean Ticket Price based on the Arrival Times

data.groupby('arrival_time')['price'].mean()
```

```python
# catplot() is a categorical plots. Used for visualizing categorical data (discrete groups) 
# with different representations like bars, boxes, or points.

sns.catplot(x = 'departure_time',
        y = 'price',
        kind = 'bar',
        palette = 'bright',
        data = data)
plt.show()
```

```python
# relplot() is relational plot. They provide a unified interface to visulaize differnt types of data while
# automatically handling subplots and legends.

sns.relplot(x = 'arrival_time',
           y = 'price',
           data = data,
           col = 'departure_time',
           kind = 'line')

plt.show()
```
